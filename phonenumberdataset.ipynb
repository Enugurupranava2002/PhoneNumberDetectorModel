```python
phone_number_dataset = pd.DataFrame(index=range(40000),columns=["image_id", "phone_number"])
```

```python
dataset_length = 40000
for i in range(dataset_length):
    random_indicies = [randint(0, dataset_length) for i in range(10)]
    # print(f"random indicies:\n{random_indicies}")
    phone_number = str(''.join(str(num) for num in [labels.loc[i, 'label'] for i in random_indicies]))
    # print(f"phone number generated:\n{phone_number}")
    images_id_list = [f"data\mnistImages\{id}.png" for id in random_indicies]
    # print(f"images id list:\n{images_id_list}")
    img1 = Image.open(images_id_list.pop(0))
    for im in images_id_list:
        img2 = Image.new('RGB', (img1.width + Image.open(im).width, 28), (0,0,0))
        img2.paste(img1, (0, 0))
        img2.paste(Image.open(im), (img1.width, 0))
        img1 = img2
    img1.save(f"data\phone_number_images\phoneNumberImage{i}.png", "PNG")
    phone_number_dataset.loc[i,'image_id'] = f"phoneNumberImage{i}.png"
    phone_number_dataset.loc[i, 'phone_number'] = phone_number
    if(i % 500 == 0):
        print("-", end="")
```
